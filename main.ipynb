# Importação das Bibliotecas Python

```python
# Biblioteca usada para manipular e analisar Dados.
import pandas as pd

# Biblioteca que auxilia na manipulação dos Dados.
import numpy as np

# Bibliotecas usadas na visualização dos Dados.
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set_theme()
```

# Importação dos Dados

```python
# Lendo um arquivo CSV usando a biblioteca Pandas
data = pd.read_csv('data.csv')
```

# Preenchendo os Dados Faltantes

```python
data['ShortPassing'] = data['ShortPassing'].fillna(data['ShortPassing'].mean())
data['Volleys'] = data['Volleys'].fillna(data['Volleys'].mean())
data['Dribbling'] = data['Dribbling'].fillna(data['Dribbling'].mean())
data['Curve'] = data['Curve'].fillna(data['Curve'].mean())
data['FKAccuracy'] = data['FKAccuracy'].fillna(data['FKAccuracy'].mean())
data['LongPassing'] = data['LongPassing'].fillna(data['LongPassing'].mean())
data['BallControl'] = data['BallControl'].fillna(data['BallControl'].mean())
data['HeadingAccuracy'] = data['HeadingAccuracy'].fillna(data['HeadingAccuracy'].mean())
data['Finishing'] = data['Finishing'].fillna(data['Finishing'].mean())
data['Crossing'] = data['Crossing'].fillna(data['Crossing'].mean())
data['Weight'] = data['Weight'].fillna('200lbs')
data['Contract Valid Until'] = data['Contract Valid Until'].fillna(2019)
data['Height'] = data['Height'].fillna("5'11")
data['Loaned From'] = data['Loaned From'].fillna('None')
data['Joined'] = data['Joined'].fillna('Jul 1, 2018')
data['Jersey Number'] = data['Jersey Number'].fillna(8)
data['Body Type'] = data['Body Type'].fillna('Normal')
data['Position'] = data['Position'].fillna('ST')
data['Club'] = data['Club'].fillna('No Club')
data['Work Rate'] = data['Work Rate'].fillna('Medium/ Medium')
data['Skill Moves'] = data['Skill Moves'].fillna(data['Skill Moves'].median())
data['Weak Foot'] = data['Weak Foot'].fillna(3)
data['Preferred Foot'] = data['Preferred Foot'].fillna('Right')
data['International Reputation'] = data['International Reputation'].fillna(1)
data['Wage'] = data['Wage'].fillna('€200K')
```

```python
def defending(data):
    columns = ['Marking', 'StandingTackle', 'SlidingTackle']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None  # Ou substitua por 0 ou outro valor padrão
        return int(round(mean_value))
    return None

def general(data):
    columns = ['HeadingAccuracy', 'Dribbling', 'Curve', 'BallControl']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None

def mental(data):
    columns = ['Aggression', 'Interceptions', 'Positioning', 'Vision', 'Composure']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None

def passing(data):
    columns = ['Crossing', 'ShortPassing', 'LongPassing']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None

def mobility(data):
    columns = ['Acceleration', 'SprintSpeed', 'Agility', 'Reactions']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None

def power(data):
    columns = ['Balance', 'Jumping', 'Stamina', 'Strength']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None

def rating(data):
    columns = ['Potential', 'Overall']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None

def shooting(data):
    columns = ['Finishing', 'Volleys', 'FKAccuracy', 'ShotPower', 'LongShots', 'Penalties']
    if all(col in data.index for col in columns):
        mean_value = data[columns].mean(skipna=True)
        if np.isnan(mean_value):
            return None
        return int(round(mean_value))
    return None
```

```python
data['Defending'] = data.apply(defending, axis = 1)
data['General'] = data.apply(general, axis = 1)
data['Mental'] = data.apply(mental, axis = 1)
data['Passing'] = data.apply(passing, axis = 1)
data['Mobility'] = data.apply(mobility, axis = 1)
data['Power'] = data.apply(power, axis = 1)
data['Rating'] = data.apply(rating, axis = 1)
data['Shooting'] = data.apply(shooting, axis = 1)
```

```python
players = data[['Name','Defending','General','Mental','Passing',
                'Mobility','Power','Rating','Shooting','Flag','Age', 
                'Preferred Foot', 'Nationality', 'Photo', 'Club Logo', 
                'Club', 'Position', 'Wage', 'Weight']]

players.head()
```

# Visualização de Dados

```python

# Comparação entre Destros e Canhotos

plt.rcParams['figure.figsize'] = (5, 5)
sns.countplot(x= players['Preferred Foot'], palette= 'pastel')
plt.title('Comparação entre Destros e Canhotos')
plt.xlabel('Pé')
plt.ylabel('Qtd. de Jogadores')
plt.show()
```

```python
# Grafico de Pizza para demonstrar a proporção das reputações Internacionais
labels = ['1', '2', '3', '4', '5']
sizes = data['International Reputation'].value_counts()
colors = plt.cm.copper(np.linspace(0, 1, 5))

plt.rcParams['figure.figsize'] = (5, 5)
plt.pie(sizes, labels=labels, colors=colors, shadow=False)
plt.title('Reputação Internacional')
plt.legend()
plt.show()
```

```python
# Numero de Jogadores por Posição

plt.rcParams['figure.figsize'] = (18, 8)
# plt.style.use('fivethirtyeight')
sns.countplot(x= players['Position'], palette='pastel')
plt.title('Numero de Jogadores por Posição', color='black')
plt.xlabel('Posição', color='black')
plt.ylabel('Qtd. de Jogadores', color='black')
plt.xticks(color='black')

plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15, 10)
sns.histplot(players['Wage'], color='blue')
plt.xlabel('Salario', fontsize = 16)
plt.ylabel('Nº de Jogadores', fontsize = 16)
plt.title('Distribuição do Salario dos Jogadores', fontsize = 20)
plt.xticks(rotation = 90)
plt.show()
```

```python
# Distribuição de Jogadores por país
plt.style.use('dark_background')
data['Nationality'].value_counts().head(80).plot.bar(color = 'orange', figsize = (20, 7))
plt.title('Qtd. de Jogadores por País', fontsize = 30, fontweight = 20)
plt.xlabel('País')
plt.ylabel('Qtd. de Jogadores')
plt.show()
```

```python
# Histograma com a distribuição dos pesos dos Jogadores
sns.histplot(data['Weight'], color='pink', kde=True)
plt.title('Distribuição dos Pesos dos Jogadores no FIFA 2019', fontsize=20)
plt.xlabel('Peso em Libras', fontsize=16)
plt.ylabel('Qtd. de Jogadores', fontsize=16)
plt.show()
```

```python
# Boxplot dos Salarios dos Jogadores
clubs = ['CD Leganés', 'Southampton', 'RC Celta', 'Empoli', 'Fortuna Düsseldorf', 'Manchester City',
         'Tottenham Hotspur', 'FC Barcelona', 'Juventus', 'Arsenal', 'Real Madrid']
dataClubs = data.loc[data['Club'].isin(clubs)]

# Configurando o tamanho da figura
plt.rcParams['figure.figsize'] = (15, 8)

# Criando o boxplot para os salários, definindo o 'hue' como None
ax = sns.boxplot(x='Club', y='Overall', data=dataClubs, hue=None, palette='pastel', legend=False)

# Ajustando rótulos e título
ax.set_xlabel('Clube', fontsize=9)
ax.set_ylabel('Salário dos Jogadores', fontsize=9)
ax.set_title('Distribuição dos Salários dos Jogadores em diferentes Clubes', fontsize=20)

# Rotacionando os rótulos do eixo X
plt.xticks(rotation=90)

# Exibindo o gráfico
plt.show()
```
